```python
#Q8
import matplotlib.pyplot as plt
# Domaine [-3, 3]
x = np.linspace(-3, 3, 100)

# Plot des fonctions
plt.plot(x, x**2 - 1, color='red', label='f(x) = x^2 - 1')
plt.plot(x, 3*x**3 - 2*x**2 + x + 1, color='blue', label='f(x) = 3x^3 - 2x^2 + x + 1')

# Légende et titre
plt.legend()
plt.title('Graphique de deux fonctions')

# Affichage du plot
plt.show()
```

```python
#Q9
import matplotlib.pyplot as plt
# nombre de nombres premiers dans chaque plage
nombres_premiers = [25, 143, 1061]
# couleurs pour chaque plage
couleurs = ['blue', 'red', 'green']
# légende pour chaque plage
legendes = ['0-100', '101-1000', '1001-10000']
# créer un diagramme circulaire
plt.pie(nombres_premiers, colors=couleurs, labels=legendes, autopct='%1.1f%%')
# ajouter un titre
plt.title('Quantité de nombres premiers dans différentes plages')
# afficher le diagramme
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calcul des nombres premiers entre 0 et 1000
nombres_premiers = np.arange(1001)
nombres_premiers[0] = 0
for n in range(2, int(np.sqrt(1000))+1):
    nombres_premiers[(nombres_premiers != n) & (nombres_premiers % n == 0)] = 0
nombres_premiers = nombres_premiers[nombres_premiers != 0]

# Création des intervalles pour l'histogramme
intervalles = np.arange(0, 1001, 100)

# Comptage du nombre de nombres premiers dans chaque intervalle
nombres_premiers_intervalles, _ = np.histogram(nombres_premiers, bins=intervalles)

# Création du plot
fig, ax = plt.subplots()
ax.bar(intervalles[:-1], nombres_premiers_intervalles, width=100)

# Définition des étiquettes pour les graduations en bas du plot
etiquettes = [f"{i}-{i+99}" for i in range(0, 1000, 100)]
ax.set_xticklabels(etiquettes)

# Affichage du plot
plt.show()
```

```python
#Q11
import pandas as pd
import numpy as np
# Création du DataFrame
df = pd.DataFrame({'continent': ['Asie', 'AmNord', 'AmNord', 'Europe', 'Europe', 'Europe', 'Asie', 'Europe', 'Europe'],
                   'country': ['Chine', 'Etats-Unis', 'Canada', 'Pologne', 'Roumanie', 'Italie', 'Inde', 'Allemagne', 'Ukraine'],
                   'PIB': np.random.randint(1, 12, 9),
                   'Membre': np.random.choice(['O', 'N'], 9)})

# Regroupement des données par continent et somme du PIB pour chaque groupe
pib_par_continent = df.groupby('continent')['PIB'].sum()

# Filtrage des données pour ne conserver que les membres des Nations Unies Virtuelles
pib_membres = df[df['Membre'] == 'O']

# Regroupement des données filtrées par continent et somme du PIB pour chaque groupe
pib_membres_par_continent = pib_membres.groupby('continent')['PIB'].sum()

# Affichage du résultat
print(pib_membres_par_continent)
```

```python
#Q12
import pandas as pd
import numpy as np
# Création du DataFrame
df = pd.DataFrame({'continent' : ['Asie','AmNord','AmNord','Europe','Europe', 'Europe','Asie',
                                  'Europe', 'Europe'],
                   'country' : ['Chine', 'Etats-Unis', 'Canada', 'Pologne', 'Roumanie', 'Italie', 'Inde',
                                'Allemagne', 'Ukraine'],
                   'PIB' : np.random.randint(1, 12 , 9),
                   'Membre' : np.random.choice(['O', 'N'], 9)})

# Création de la table pivotante
pivot_table = pd.pivot_table(df, values='PIB', index='continent', columns='Membre', aggfunc=np.sum, fill_value=0)

# Création du nouveau DataFrame
new_df = pd.DataFrame({'PIB des membres' : pivot_table['O'], 'PIB des non-membres' : pivot_table['N']})

# Affichage du nouveau DataFrame
print(new_df)
```

```python
import pandas as pd

df = pd.DataFrame({'Ville':['limoges','limoges','limoges','limoges','limoges','brive','brive',
                            'brive','brive','brive','gueret','gueret','gueret'],
                   'Zone':[1,1,2,2,1,1,2,1,2,3,1,2,3],
                   'Sales':[10,16,21,37,20,40,37,51,17,24,11,30,20]})

resultat = df.groupby(['Ville','Zone'])['Sales'].sum().reset_index().sort_values(['Ville','Zone'])
print(resultat)
```

```python
import pandas as pd
#Q15
# Création du DataFrame
df = pd.DataFrame({
    "Nom": ["John", "Steve", "Martin", "Martin", "Steve", "Martin"],
    "Ville": ["Limoges", "Limoges", "Brive", "Limoges", "Limoges", "Brive"],
    "Ventes": [130, 121, 126, 144, 120, 110]
})

# Ajout d'une colonne avec la somme des ventes pour chaque paire (Nom, Ville)
df["Ventes_totales"] = df.groupby(["Nom", "Ville"])["Ventes"].transform("sum")

# Affichage du DataFrame résultant
print(df)
```
